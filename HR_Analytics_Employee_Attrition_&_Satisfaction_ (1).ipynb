```python
corr_with_attrition = df.corr(numeric_only=True)['Attrition'].sort_values(ascending=False)
print(corr_with_attrition)
```

```python
top_corr = corr_with_attrition.drop('Attrition').head(35)  # top 10 excluding itself

plt.figure(figsize=(8,6))
sns.barplot(x=top_corr.values, y=top_corr.index, legend=False,palette='coolwarm')
plt.title("Top 10 Factors Correlated with Employee Attrition", fontsize=14, fontweight='bold')
plt.xlabel("Correlation with Attrition")
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
correlation=df.corr(numeric_only=True)
correlation["Attrition"]>0.5
correlation
```

**Key Findings** 
1. Attrition Rate : 16%
2. Employees in early tenure ( 0-2 years ) show highest attrition rates meaning gaps in onboarding and retention .
3. Overtime is a driver for attrition indicating a worload inbalance .
4. Longer gaps between promotion is also a highest attrition factor.
5. Employees with lower job satisfaction are also likely to exit .
