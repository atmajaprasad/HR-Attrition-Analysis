```python
plt.figure(figsize=(15,15))
corr = df.corr(numeric_only=True)
mask = (corr.abs() > 0.5)
sns.heatmap(corr.where(mask), annot=True, cmap="coolwarm", fmt=".2f")
plt.grid(True)
plt.title("Significant Correlations Only (|r| > 0.5)", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
corr_with_attrition = df.corr(numeric_only=True)['Attrition'].sort_values(ascending=False)
print(corr_with_attrition)
```

```python
top_corr = corr_with_attrition.drop('Attrition').head(35)  # top 10 excluding itself

plt.figure(figsize=(8,6))
sns.barplot(x=top_corr.values, y=top_corr.index, legend=False,palette='coolwarm')
plt.title("Top 10 Factors Correlated with Employee Attrition", fontsize=14, fontweight='bold')
plt.xlabel("Correlation with Attrition")
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
